```python
def _safe_mask(invalid_mask):
    valid = ~invalid_mask                                      # [B, N] True=valid
    all_invalid = ~valid.any(dim=-1, keepdim=True) 
    print(all_invalid)             # [B, 1] True when whole row is padding
    return valid | all_invalid                                  # unmask full row for all-padding elements


invalid_mask = torch.tensor([[True, True, True], [False, True, True]], device="cuda")
print(invalid_mask.shape)
print(_safe_mask(invalid_mask))
```
